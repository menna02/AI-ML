#### Showing the image before and after Detection

```python
def show_images(img_path1, img_path2):
    img1 = cv2.imread(img_path1)
    img2 = cv2.imread(img_path2)
    plt.figure(figsize=(5, 5))
    
    plt.subplot(1,2,1)
    plt.title('Test Image Before Detection', fontsize=15, color='red')
    plt.imshow(img1)
    plt.axis(False)
    plt.tight_layout()
    
    plt.subplot(1,2,2)
    plt.title('Test Image After Detection', fontsize=15, color='blue', pad=20)
    plt.imshow(img2)
    plt.axis(False)
    plt.tight_layout()
    
    plt.show()
    
    
```

```python
img_path='/kaggle/input/cars-detection/Cars Detection/test/images/00dea1edf14f09ab_jpg.rf.3f17c8790a68659d03b1939a59ccda80.jpg'
detected_image_path = '/kaggle/working/runs/detect/predict3/00dea1edf14f09ab_jpg.rf.3f17c8790a68659d03b1939a59ccda80.jpg'
show_images(img_path, detected_image_path)
```
